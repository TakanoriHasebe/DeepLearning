```python
# ライブラリのロード
import seaborn as sns
import matplotlib.pyplot as plt
from zodbpickle import pickle
sns.set_style('whitegrid')
%matplotlib inline
```

```python
with open('acc_list_weight_random_no_batch.pkl', 'rb')\
     as f: random_no_batch = pickle.Unpickler(f).load()
with open('acc_list_weight_random_with_batch.pkl', 'rb')\
     as f: random_with_batch = pickle.Unpickler(f).load()
with open('acc_list_weight_xavier_no_batch.pkl', 'rb')\
     as f: xavier_no_batch = pickle.Unpickler(f).load()
with open('acc_list_weight_xavier_no_batch.pkl', 'rb')\
     as f: xavier_with_batch = pickle.Unpickler(f).load()
```

```python
len(random_no_batch)
```

```python
x = range(0, 50)
y = random_no_batch
plt.plot(x, y)
```

```python
x = range(0, 50)
y = random_with_batch
plt.plot(x, y)
```

```python
y = xavier_no_batch
plt.plot(x, y)
```

```python
y = xavier_with_batch
plt.plot(x, y)
```

```python
with open('elapsed_time_xavier_no_batch.pkl', 'rb')\
     as f: time_xavier_no_batch = pickle.Unpickler(f).load()
with open('elapsed_time_xavier_with_batch.pkl', 'rb')\
     as f: time_xavier_with_batch = pickle.Unpickler(f).load()
```

```python
# 自分でPythonで作成した初期重みにxavierを用いた時にかかった時間
time_xavier_no_batch / 60
```

```python
time_xavier_with_batch / 60
```

### <font color='red'><考察></font>
もう少し色々な初期の重みを試す必要がある。<br>
時間はBatch Normalizationを加えたほうがかかった。

```python
with open('elapsed_time_chainer.pkl', 'rb')\
     as f: time_chainer = pickle.Unpickler(f).load()
```

```python
time_chainer / 60
```
